<h2>Part 1 Analysis of iterative , recursion, tail and non tail algorithms</h2>
<h3> Section 1 </h3>
<blockquote> Iterative approach </blockquote>
    <p> In Iterative approach we use a for loop to create the factorial series <p>
    <p>AlGORITHM</p>
    <ol><li>Define function called factorial_iterative(x)</li>
        <li>If we input 1 return directly 1</li>
        <li>If number greater than 1 is sent as parameter make a for loop which will iterate from 1 to given number</li>
        <li>Factorial calculated in variable called fact as fact = fact*number </li>
        <li>Finally return the factorial calculated</li>
        
        
        


```python
def factorial_iterative(x):
    if (x==1):
        return 1
    else:
        fact = 1
    for z in range(1,x+1):
        fact=fact*z
    return fact
```

<h3> Section 1.1 -> Iterative process Analysis </h3>
   <blockquote>Analysis of program</blockquote>
   <code>
   def factorial_iterative(x): --> 2 operations   > function call and passing data
    if (x==1):                 --> 2 operations   > if condition , comparision
        return 1               --> 1 operation 
    else:
        fact = 1               --> 1 operation    > assignment 
    for z in range(1,x+1):     --> 2n operation   > for loop and assignment per loop
        fact=fact*z            --> 2n operations  > assignment , multiplication per loop
    return fact                --> 1 opeartion
    </code>
    <blockquote> O(4n+4) --> O(n)

<h3>Section 1.2 -> Time vs N plot</h3>
<blockquote>Test script</blockquote>
<p> In order to plot the time complexity a script is made<br>
    the script runs the factorial function with increaseing size of N<br>

<h3> Section 2</h3>
<blockquote>Section 2 -> Recursion normal (non-tail) </blockquote>
    <p> In this approach we use the normal recursive calls to generate factorials <p>
    <p>AlGORITHM</p>
    <ol><li>Define function called factorial_normal(x)</li>
        <li>If we input 1 return directly 1</li>
        <li>If number greater than 1 is sent as parameter make a call to factorial_normal with parameter being (x-1) the recursion occurs in this step as we are calling the same function again with modified parameter </li>
        <li>the recursive calls happen till parameter becomes 1 at that point the first if condition activates and returns 1 </li>
        <li>Now the stacks which contain numbers pop one by one as  x*factorial_normal(x-1) which creates an expression like this -> x*(x-1)*x(x-2) upto 1 </li>
    <li> return the exprssion which will be the factorial result</li>
        

```python
def factorial_normal(x):
    if (x==1):
        return 1
    else:
        return x*factorial_normal(x-1)
```

<h3> Section 2.1 -> Recursion normal (non-tail) Analysis </h3>
   <blockquote>Analysis of program</blockquote>
   <code>
   def factorial_normal(x):             --> 2 operations   > function call and passing data
    if (x==1):                          --> 2 operations   > if condition , comparision
        return 1                        --> 1 operation 
    else:
        return x*factorial_normal(x-1)  --> 3 operation    > multiplication, function call which has assignment PER ACTIVATION
    </code>
    <blockquote> O(5n-2) --> O(n)

<h3> Section 3 </h3>
<blockquote> Recursion tail approach </blockquote>
    <p> In tail recursion method we directly pass result from return statement. No other operations will take place apart from function call and passing of parameters <p>
    <p>AlGORITHM</p>
    <ol><li>Define function called factorial_tail(x,fact) unlike the previous function this has 2 parameters one called an accumilator and other which will have actual result initially set to 1</li>
        <li>If we input 1 or 0 return  1 and end program</li>
        <li>If number greater than 1 is sent as parameter(accumilator) a recursive call is made with parameters being number*accumilator,accumilator-1 </li>
        <li>Here we can see that no operations happen in return statement and all math operations occure inside parameter  </li>
        <li>Finally return the factorial calculated which will happen when accumilator goes to 1 and we return the other parameter</li>
        
        
        



```python
def factorial_tail(x,fact):
    if fact == 1 or fact ==0:
        return x
    else:
        return factorial_tail(x*fact,fact-1)
```

<h3> Section 3.1 -> Tail recursion Analysis </h3>
   <blockquote>Analysis of program</blockquote>
   <code>
   def factorial_tail(x,fact):                    --> 2 operations     > function call and passing data
        if fact == 1 or fact ==0:                 --> 2n operations    > if condition , comparision
            return x                              --> 1 operation 
        else:
        return factorial_tail(x*fact,fact-1)      --> 2(n-1) opeartions
    </code>
    <blockquote> O(4n+1) --> O(n)

<h3>Section 4.1 Analysis of binary search algorithms</h3>
<blockquote> Array generation </blockquote>
<p> In this section we will generate an array of size 150 with each cell storing factorial from 1 to 150. The importtant thing to note is that the <b>ARRAY IS SORTED</b>. This array will be used for doing the doing the binary sort comparision

```python
factorial_array = [factorial_iterative(x) for x in range (1,150)]
```

<h3>Section 4.2 Analysis of binary search algorithms</h3>
    <p>AlGORITHM</p>
    <ol><li>Define function called binary_search(arr, num, first=0, last=len(factorial_array)) where arr is sorted array. num is the number to be found</li>
        <li>Repeat Iteration untill 'first' greater than 'last' </li>
        <li>Find mid point of array and see if the number we search is there </li>
        <li>If the element is not found remove half of the array based on mid point value and where our number is to be expected</li>
        <li> Repeat the process untill we find element we search for or untill the array becomes to small to check </li>
        

```python
def binary_search(arr, num, first=0, last=len(factorial_array)):
    while first < last:
        mid = (first + last) // 2
        if arr[mid] == num:
            return True
        else:
            if arr[mid] < num:
                first = mid + 1
            else:
                last = mid
    else:
        return False            
```

<h3>Section 4.3 -> Time vs N plot unit test</h3>
<blockquote>Test script</blockquote>
<p> In order to plot the time complexity a script is made<br>
    the script runs the binary_search function with increaseing size of N<br>
    In order to make the curve as smooth as possible the process is repeated 1000 times and answer is averaged and finally a plot is made with x- axis being input size (N) and y-axis being the time taken to find worst case element search</p>



```python
import time
import matplotlib.pyplot as plt
def timer_averaged_binary_search(fx,x):
    global av_plt_binary_search
    x_axis = [x for x in range (0,x)]
    y_axis = [0 for x in range (0,x)]
    avg_plot=[0 for x in range (0,x)]
    
    
    for z in range (1000):
        for k in range (1,x):
            factorial_array = [factorial_iterative(x) for x in range (0,k+1)]
            last_value_array = factorial_array[k]
            start=time.time()
            fx(factorial_array,last_value_array,0,len(factorial_array))
            end=time.time()
            avg_plot[k] = avg_plot[k] + (end - start )
        
    
    av_plt_binary_search = [s/1000 for s in avg_plot]
    plt.title("Binary search iterative")
    plt.xlabel('Number of inputs (N)')
    plt.ylabel('time taken for execution in seconds')
    plt.plot(x_axis,av_plt_binary_search)
    plt.show()
```

<h3>Section 4.5 -> Analysis of binary search algorithms</h3>
    <p>AlGORITHM</p>
    <ol><li>Define function called linear_search(arr,num) where arr is sorted array. num is the number to be found</li>
        <li>Use a for loop to go through each index of array and use if condition to check if elemnts we go through correspond to elemnt we are searching for. </li>
        <li>If element is found return True else return False </li>
        

```python
def linear_search(arr,num):
    for x in range (0,len(arr)):
        if arr[x] == num:
            return True
    else:
        return False
```
